```python
df=pd.read_csv('Downloads/Cars93.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df.isna().sum().plot(kind='bar')
```

```python
# filling null values
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isna().sum()
```

```python
X=df.drop(labels=['Price','EngineSize','Horsepower','Type'],axis=1)
Y=df[['AirBags']]
```

# separate categorical and continues features from input features

```python
cat=[]
con=[]
for i in X.columns:
    if X[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
Xcat=X[cat]
Xcon=X[con]
```

```python
X
```

# Preprocessing
Standardisation

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
Xcon=pd.DataFrame(ss.fit_transform(Xcon),columns=con)
```

```python
Xcon
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in Xcat.columns:
    Xcat[i]=le.fit_transform(Xcat[i])
```

```python
Xcat
```

```python
X=Xcon.join(Xcat)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
model=lr.fit(xtrain,ytrain)
```

```python
tr_pred=model.predict(xtrain)
```

```python
ts_pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
tr_score=accuracy_score(ytrain,tr_pred)
```

```python
ts_score=accuracy_score(ytest,ts_pred)
```

```python
print(tr_score)
```
